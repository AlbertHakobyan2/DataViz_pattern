```python
def computing_iqr_stats(df, column):
    stats = df.groupby("Labels")[column].describe()[["25%", "75%"]]
    stats["IQR"] = stats["75%"] - stats["25%"]
    stats["Lower Bound"] = stats["25%"] - 1.5 * stats["IQR"]
    stats["Upper Bound"] = stats["75%"] + 1.5 * stats["IQR"]
    return stats

iqr_stats_A = computing_iqr_stats(excel, "A")
iqr_stats_B = computing_iqr_stats(excel, "B")

print("IQR Statistics for Column A:\n", iqr_stats_A)
print("\nIQR Statistics for Column B:\n", iqr_stats_B)
```

```python
from scipy.stats import ks_2samp

midpoint = len(excel) // 2
# Now, splitting into 2 equal halves + storing them into early and late portions:
excel_early = excel.iloc[:midpoint]
excel_late = excel.iloc[midpoint:]

# Performing Kolmogorov–Smirnov test on each feature. If p-value of any feature is <0.05 we will conclude an existence of concept drift
ks_results = {"A": ks_2samp(excel_early["A"], excel_late["A"]).pvalue, "B": ks_2samp(excel_early["B"], excel_late["B"]).pvalue,
              "Labels": ks_2samp(excel_early["Labels"], excel_late["Labels"]).pvalue}

print(ks_results)
```

A ($p-value ≈ 3.42 * 10^{-06}$) → Very small p-value, meaning the distribution of column A has significantly changed between the first and second halves of the dataset. This means that there is a concept drift in A.

B (p-value ≈ 0.78) → High p-value, meaning the distribution of B has remained mostly the same. 
Basically, no significant drift detected here.

Labels (p-value = 1.0), meaning perfectly identical distributions in both halves, n shift over time.


```python
excel["A_bins"] = pd.qcut(excel["A"], q=20) #
label_drift = excel.groupby(["A_bins", "Segment"])["Labels"].value_counts().unstack()

label_drift.plot(kind="bar", stacked=True, figsize=(12, 6))
plt.title("Distribution of Labels Across A Bins (Early vs Late)")
plt.ylabel("Probabilities")
plt.show()
```

## Early vs Late Distribution:

### Early Phase (Left side):

- Higher concentration of labels 2 (green) and 3 (red)
- Lower presence of labels 0 (blue) and 1 (orange)
- The pattern suggests a preference for higher-value labels

### Late Phase (Right side):

Dramatic increase in labels 0 (blue) and 1 (orange)
Significant decrease in labels 2 and 3
Shows a clear shift towards lower-value labels

#### Regions of Pronounced Drift:
- The most pronounced drift appears in three regions:
    - Initial Transition (around A_bin ≈ 0.2-0.3):
    - Sharp decrease in label 3 (red)
    - Increase in label 2 (green)

#### Middle Transition (around A_bin ≈ 0.5-0.6):
- Almost no change

#### Late Transition (around A_bin ≈ 0.8-0.9):
- Rapid increase in label 0 (blue)
- Almost complete disappearance of label 3

#### So I can conlude the following: This data has a complex form of concept drift where:
- The drift is primarily evident in feature A's relationship with the labels, not with the B column.
- The drift appears to be gradual but with distinct transition phases

According to the article, this **pattern** suggests a "virtual concept drift" where the underlying relationship between features and labels changes while maintaining the same marginal distributions of the labels. This is interesting because it indicates a change in the decision boundaries while preserving the overall class balance (if I get this correct).
